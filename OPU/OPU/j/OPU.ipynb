```python
def find_nearest_index(array, value):
  array = np.asarray(array)
  idx = (np.abs(array - value)).argmin()
  return idx
```

```python
def plot(data, xmin=None, xmax=None, fitfunc=None, p0=None, **kwargs):
  fig = plt.figure(figsize=(12,8), dpi=112)
  
  x = data[[data.columns[0]]].values.ravel()
  y = data[[data.columns[1]]].values.ravel()
  
  if xmin != None and xmax != None:
    
    xmin = find_nearest_index(x, xmin)
    xmax = find_nearest_index(x, xmax)
    
    x = x[xmin:xmax]
    y = y[xmin:xmax]
  
  if fitfunc != None and xmin != None and xmax != None:
    from scipy.optimize import curve_fit
    
    print(fitfunc(x, 1,1).shape)
    print(np.min(x), np.max(x))
    
    popt, pcov = curve_fit(fitfunc, x, y)
    
    xx = np.linspace(xmin, xmax, 100)
    plt.plot(xx, fitfunc(xx, *popt), 'r-', label='Fitted function')
  
  plt.plot(x, y, **kwargs)
```

# t2

```python
t21 = pd.read_csv("t2/better/test1.csv", comment="#") #, comments="#",unpack=True, delimiter=",")
t22 = pd.read_csv("t2/better/test2.csv", comment="#") #, comments="#",unpack=True, delimiter=",")
```

```python
plot(t21)
```

```python
from scipy.optimize import curve_fit

x = t21[[t21.columns[0]]].values.ravel()
y = t21[[t21.columns[1]]].values.ravel()

xmin = 0.00875
xmax = 0.0096

xmin_i = find_nearest_index(x, xmin)
xmax_i = find_nearest_index(x, xmax)

x=x[xmin_i:xmax_i]
y=y[xmin_i:xmax_i]

xx = np.linspace(xmin, xmax, 100)


def ff(x, x1, x2, tau):
  return x1 + x2*np.exp(-x/tau).ravel()

p0 = (-0.05, 0.1, 0.0005)
popt, pcov = curve_fit(ff, x, y, p0=p0)
print(popt)

fig = plt.figure(figsize=(12,8), dpi=112)
plt.plot(x ,y, color='c', marker="+", linestyle='None')
plt.plot(xx, ff(xx, *popt), color='r', marker='None', linestyle='-')
```

```python
def latexRowToArray(s):
  nums = s.split("&")
  for k in range(len(nums)):
    nums[k] = ''.join([i for i in nums[k] if i.isdigit() or i in ['.']])
    nums[k] = float(nums[k])
  return np.array(nums)

def arrayToLatexRow(a, nlen=5):
  ns = []
  for n in a:
    ns.append("${0}$".format(round(n, nlen)))
    
  ns = ' & '.join(ns)
  ns = ns + " \\ \\hline"
  return ns
```
